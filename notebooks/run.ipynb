```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
import math
from pathlib import Path
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from datasets import load_dataset
from tokenizers import Tokenizer
from tokenizers.models import WordLevel
from tokenizers.trainers import WordLevelTrainer
from tokenizers.pre_tokenizers import Whitespace
import pandas as pd
import matplotlib.pyplot as plt  # for plotting

# ---------------------------
# Config definition
# ---------------------------
def get_config():
    return {
        "batch_size": 156,  
        "num_epochs": 10,         # Set epochs to 10
        "lr": 1e-4,
        "seq_len": 256,
        "d_model": 256,
        "datasource": "cfilt/iitb-english-hindi", 
        "lang_src": "en",
        "lang_tgt": "hi",
        "model_folder": "weights",
        "model_basename": "tmodel_",
        "preload": None,  # or 'latest' or "00" etc. if you want to resume
        "tokenizer_file": "tokenizer_{0}.json",
        "experiment_name": "runs/iitb_tmodel_en_hi"
    }

def get_weights_file_path(config, epoch: str):
    model_folder = f"{config['datasource'].replace('/', '_')}_{config['model_folder']}"
    model_filename = f"{config['model_basename']}{epoch}.pt"
    return str(Path('.') / model_folder / model_filename)

def latest_weights_file_path(config):
    model_folder = f"{config['datasource'].replace('/', '_')}_{config['model_folder']}"
    model_filename = f"{config['model_basename']}*"
    weights_files = list(Path(model_folder).glob(model_filename))
    if not weights_files:
        return None
    weights_files.sort()
    return str(weights_files[-1])

# ---------------------------
# Dataset code
# ---------------------------
class BilingualDataset(Dataset):
    def __init__(self, ds, tokenizer_src, tokenizer_tgt, src_lang, tgt_lang, seq_len):
        super().__init__()
        self.ds = ds
        self.seq_len = seq_len
        self.tokenizer_src = tokenizer_src
        self.tokenizer_tgt = tokenizer_tgt
        self.src_lang = src_lang
        self.tgt_lang = tgt_lang

        self.sos_token = torch.tensor([tokenizer_tgt.token_to_id("[SOS]")], dtype=torch.long)
        self.eos_token = torch.tensor([tokenizer_tgt.token_to_id("[EOS]")], dtype=torch.long)
        self.pad_token = torch.tensor([tokenizer_tgt.token_to_id("[PAD]")], dtype=torch.long)

    def __len__(self):
        return len(self.ds)

    def __getitem__(self, idx):
        row = self.ds[idx]
        src_text = row["translation"][self.src_lang]
        tgt_text = row["translation"][self.tgt_lang]

        enc_input_tokens = self.tokenizer_src.encode(src_text).ids
        dec_input_tokens = self.tokenizer_tgt.encode(tgt_text).ids

        # Truncate
        enc_input_tokens = enc_input_tokens[: self.seq_len - 2]
        dec_input_tokens = dec_input_tokens[: self.seq_len - 1]

        # Padding calculation
        enc_pad_len = self.seq_len - len(enc_input_tokens) - 2
        dec_pad_len = self.seq_len - len(dec_input_tokens) - 1

        # Encoder input
        encoder_input = torch.cat([
            self.sos_token,
            torch.tensor(enc_input_tokens, dtype=torch.long),
            self.eos_token,
            torch.tensor([self.pad_token.item()] * enc_pad_len, dtype=torch.long),
        ], dim=0)

        # Decoder input
        decoder_input = torch.cat([
            self.sos_token,
            torch.tensor(dec_input_tokens, dtype=torch.long),
            torch.tensor([self.pad_token.item()] * dec_pad_len, dtype=torch.long),
        ], dim=0)

        # Target (label)
        label = torch.cat([
            torch.tensor(dec_input_tokens, dtype=torch.long),
            self.eos_token,
            torch.tensor([self.pad_token.item()] * dec_pad_len, dtype=torch.long),
        ], dim=0)

        enc_mask = (encoder_input != self.pad_token).unsqueeze(0).unsqueeze(0).int()
        dec_mask = (decoder_input != self.pad_token).unsqueeze(0).int() & causal_mask(decoder_input.size(0))

        return {
            "encoder_input": encoder_input,
            "decoder_input": decoder_input,
            "encoder_mask": enc_mask,
            "decoder_mask": dec_mask,
            "label": label,
            "src_text": src_text,
            "tgt_text": tgt_text
        }

def causal_mask(size):
    mask = torch.triu(torch.ones((1, size, size)), diagonal=1).type(torch.int)
    return mask == 0

def get_all_sentences(ds, lang):
    for item in ds:
        yield item["translation"][lang]

def get_or_build_tokenizer(config, ds, lang):
    tokenizer_path = Path(config["tokenizer_file"].format(lang))
    if not tokenizer_path.exists():
        tokenizer = Tokenizer(WordLevel(unk_token="[UNK]"))
        tokenizer.pre_tokenizer = Whitespace()
        trainer = WordLevelTrainer(special_tokens=["[UNK]", "[PAD]", "[SOS]", "[EOS]"], min_frequency=2)
        tokenizer.train_from_iterator(get_all_sentences(ds, lang), trainer)
        tokenizer.save(str(tokenizer_path))
        print(f"[INFO] Tokenizer saved to {tokenizer_path}")
    else:
        tokenizer = Tokenizer.from_file(str(tokenizer_path))
        print(f"[INFO] Tokenizer loaded from {tokenizer_path}")
    return tokenizer

def get_ds(config):
    dataset = load_dataset("cfilt/iitb-english-hindi")

    # Use train split for training
    train_ds_raw = dataset["train"]
    val_ds_raw = dataset["validation"]

    print(f"[INFO] Train dataset size: {len(train_ds_raw)}")
    print(f"[INFO] Validation dataset size: {len(val_ds_raw)}")

    # Print a small sample from the training set
    sample_count = 3
    print("[INFO] Sample from training set:")
    for i in range(sample_count):
        print(f"  {i+1}. EN: {train_ds_raw[i]['translation']['en']}")
        print(f"     HI: {train_ds_raw[i]['translation']['hi']}")
        print("     ---")

    tokenizer_src = get_or_build_tokenizer(config, train_ds_raw, config["lang_src"])
    tokenizer_tgt = get_or_build_tokenizer(config, train_ds_raw, config["lang_tgt"])

    train_ds = BilingualDataset(
        train_ds_raw,
        tokenizer_src,
        tokenizer_tgt,
        config["lang_src"],
        config["lang_tgt"],
        config["seq_len"]
    )
    val_ds = BilingualDataset(
        val_ds_raw,
        tokenizer_src,
        tokenizer_tgt,
        config["lang_src"],
        config["lang_tgt"],
        config["seq_len"]
    )

    train_dataloader = DataLoader(train_ds, batch_size=config["batch_size"], shuffle=True, pin_memory=True)
    val_dataloader = DataLoader(val_ds, batch_size=1, shuffle=True, pin_memory=True)

    return train_dataloader, val_dataloader, tokenizer_src, tokenizer_tgt


# ---------------------------
# Model code
# ---------------------------
class InputEmbedding(nn.Module):
    def __init__(self, d_model:int, vocab_size:int):
        super().__init__()
        self.d_model = d_model
        self.embedding = nn.Embedding(vocab_size, d_model)

    def forward(self, x):
        return self.embedding(x.long()) * math.sqrt(self.d_model)

class PositionalEncoding(nn.Module):
    def __init__(self, d_model:int, seq_len:int, dropout:float):
        super().__init__()
        self.dropout = nn.Dropout(dropout)
        pe = torch.zeros(seq_len, d_model)
        position = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * -(math.log(10000.0)/d_model))
        pe[:,0::2] = torch.sin(position * div_term)
        pe[:,1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        seq_len = x.size(1)
        x = x + (self.pe[:, :seq_len, :]).requires_grad_(False)
        return self.dropout(x)

class layerNormalization(nn.Module):
    def __init__(self, eps:float=1e-6):
        super().__init__()
        self.eps = eps
        self.alpha = nn.Parameter(torch.ones(1))
        self.bias = nn.Parameter(torch.zeros(1))

    def forward(self, x):
        mean = x.mean(dim=-1, keepdim=True)
        std = x.std(dim=-1, keepdim=True)
        return self.alpha * (x - mean) / (std + self.eps) + self.bias

class FeedForwardBlock(nn.Module):
    def __init__(self, d_model:int, d_ff:int, dropout:float):
        super().__init__()
        self.linear1 = nn.Linear(d_model, d_ff)
        self.dropout = nn.Dropout(dropout)
        self.linear2 = nn.Linear(d_ff, d_model)

    def forward(self, x):
        x = torch.relu(self.linear1(x))
        x = self.dropout(x)
        x = self.linear2(x)
        return x

class MultiHeadAttention(nn.Module):
    def __init__(self, d_model:int, h:int, dropout:float):
        super().__init__()
        assert d_model % h == 0
        self.d_model = d_model
        self.h = h
        self.d_k = d_model // h
        self.w_q = nn.Linear(d_model, d_model)
        self.w_k = nn.Linear(d_model, d_model)
        self.w_v = nn.Linear(d_model, d_model)
        self.w_o = nn.Linear(d_model, d_model)
        self.dropout = nn.Dropout(dropout)
        self.attention_scores = None  # To store attention weights

    @staticmethod
    def attention(query, key, value, mask, dropout):
        d_k = query.size(-1)
        scores = torch.matmul(query, key.transpose(-2, -1)) / math.sqrt(d_k)
        if mask is not None:
            scores = scores.masked_fill(mask == 0, float('-inf'))
        attn = torch.softmax(scores, dim=-1)
        if dropout is not None:
            attn = dropout(attn)
        out = torch.matmul(attn, value)
        return out, attn

    def forward(self, q, k, v, mask):
        B_q, len_q, _ = q.shape
        B_k, len_k, _ = k.shape
        B_v, len_v, _ = v.shape

        query = self.w_q(q)
        key   = self.w_k(k)
        value = self.w_v(v)

        query = query.view(B_q, len_q, self.h, self.d_k).transpose(1, 2)
        key   = key.view(B_k, len_k, self.h, self.d_k).transpose(1, 2)
        value = value.view(B_v, len_v, self.h, self.d_k).transpose(1, 2)

        x, self.attention_scores = self.attention(query, key, value, mask, self.dropout)
        x = x.transpose(1, 2).contiguous().view(B_q, len_q, self.d_model)
        return self.w_o(x)

class ResidualConnection(nn.Module):
    def __init__(self, dropout:float):
        super().__init__()
        self.dropout = nn.Dropout(dropout)
        self.norm = layerNormalization()

    def forward(self, x, sublayer):
        return x + self.dropout(sublayer(self.norm(x)))

class EncoderLayer(nn.Module):
    def __init__(self, self_attention_block, feed_forward_block, dropout):
        super().__init__()
        self.self_attention_block = self_attention_block
        self.feed_forward_block = feed_forward_block
        self.residual_connections = nn.ModuleList([ResidualConnection(dropout) for _ in range(2)])

    def forward(self, x, src_mask):
        x = self.residual_connections[0](x, lambda x: self.self_attention_block(x, x, x, src_mask))
        x = self.residual_connections[1](x, self.feed_forward_block)
        return x

class Encoder(nn.Module):
    def __init__(self, layers: nn.ModuleList):
        super().__init__()
        self.layers = layers
        self.norm = layerNormalization()

    def forward(self, x, mask):
        for layer in self.layers:
            x = layer(x, mask)
        return self.norm(x)

class DecoderBlock(nn.Module):
    def __init__(self, self_attention_block, src_attention_block, feed_forward_block, dropout):
        super().__init__()
        self.self_attention_block = self_attention_block
        self.src_attention_block = src_attention_block
        self.feed_forward_block = feed_forward_block
        self.residual_connections = nn.ModuleList([ResidualConnection(dropout) for _ in range(3)])

    def forward(self, x, encoder_output, src_mask, tgt_mask):
        x = self.residual_connections[0](x, lambda x: self.self_attention_block(x, x, x, tgt_mask))
        x = self.residual_connections[1](x, lambda x: self.src_attention_block(x, encoder_output, encoder_output, src_mask))
        x = self.residual_connections[2](x, self.feed_forward_block)
        return x

class Decoder(nn.Module):
    def __init__(self, layers: nn.ModuleList):
        super().__init__()
        self.layers = layers
        self.norm = layerNormalization()

    def forward(self, x, encoder_output, src_mask, tgt_mask):
        for layer in self.layers:
            x = layer(x, encoder_output, src_mask, tgt_mask)
        return self.norm(x)

class ProjectionLayer(nn.Module):
    def __init__(self, d_model: int, vocab_size: int):
        super().__init__()
        self.proj = nn.Linear(d_model, vocab_size)

    def forward(self, x):
        return torch.log_softmax(self.proj(x), dim=-1)

class Transformer(nn.Module):
    def __init__(self, encoder, decoder, src_embed, tgt_embed, src_pos, tgt_pos, projection_layer):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.src_embed = src_embed
        self.tgt_embed = tgt_embed
        self.src_pos = src_pos
        self.tgt_pos = tgt_pos
        self.projection_layer = projection_layer

    def encode(self, src, src_mask):
        src = self.src_embed(src)
        src = self.src_pos(src)
        return self.encoder(src, src_mask)

    def decode(self, tgt, encoder_output, src_mask, tgt_mask):
        tgt = self.tgt_embed(tgt)
        tgt = self.tgt_pos(tgt)
        return self.decoder(tgt, encoder_output, src_mask, tgt_mask)

    def project(self, x):
        return self.projection_layer(x)

def build_transformer(
    src_vocab_size: int,
    tgt_vocab_size: int,
    src_seq_len: int,
    tgt_seq_len: int,
    d_model: int = 256,
    N: int = 3,
    h: int = 4,
    dropout: float = 0.1,
    d_ff: int = 1024
) -> Transformer:
    src_embed = InputEmbedding(d_model, src_vocab_size)
    tgt_embed = InputEmbedding(d_model, tgt_vocab_size)
    src_pos = PositionalEncoding(d_model, src_seq_len, dropout)
    tgt_pos = PositionalEncoding(d_model, tgt_seq_len, dropout)

    encoder_layers = []
    for _ in range(N):
        attn_block = MultiHeadAttention(d_model, h, dropout)
        ff_block = FeedForwardBlock(d_model, d_ff, dropout)
        encoder_layers.append(EncoderLayer(attn_block, ff_block, dropout))
    encoder = Encoder(nn.ModuleList(encoder_layers))

    decoder_layers = []
    for _ in range(N):
        self_attn = MultiHeadAttention(d_model, h, dropout)
        cross_attn = MultiHeadAttention(d_model, h, dropout)
        ff_block = FeedForwardBlock(d_model, d_ff, dropout)
        decoder_layers.append(DecoderBlock(self_attn, cross_attn, ff_block, dropout))
    decoder = Decoder(nn.ModuleList(decoder_layers))

    projection_layer = ProjectionLayer(d_model, tgt_vocab_size)
    transformer = Transformer(encoder, decoder, src_embed, tgt_embed, src_pos, tgt_pos, projection_layer)

    for p in transformer.parameters():
        if p.dim() > 1:
            nn.init.xavier_uniform_(p)

    return transformer

# ---------------------------
# Helper function for plotting attention
# ---------------------------
def plot_attention(attention, tokens, title="Attention Map"):
    """
    Plots a heatmap of attention weights.
    attention: Tensor of shape (heads, seq_len, seq_len)
    tokens: List of token strings corresponding to seq_len
    """
    num_heads = attention.shape[0]
    fig, axes = plt.subplots(1, num_heads, figsize=(num_heads * 3, 3))
    if num_heads == 1:
        axes = [axes]
    for i, ax in enumerate(axes):
        attn = attention[i].cpu().detach().numpy()
        im = ax.imshow(attn, cmap="viridis")
        ax.set_xticks(range(len(tokens)))
        ax.set_yticks(range(len(tokens)))
        ax.set_xticklabels(tokens, rotation=90, fontsize=6)
        ax.set_yticklabels(tokens, fontsize=6)
        ax.set_title(f"Head {i+1}")
        fig.colorbar(im, ax=ax)
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Validation / Decoding
# ---------------------------
def greedy_decode(model, src, src_mask, tokenizer_tgt, max_len=50, device="cpu"):
    """
    Greedily decode a single example:
      src, src_mask => the encoded source
      model => Transformer model
      tokenizer_tgt => for converting predicted tokens to text
    """
    sos_idx = tokenizer_tgt.token_to_id("[SOS]")
    eos_idx = tokenizer_tgt.token_to_id("[EOS]")

    # Encode the source
    with torch.no_grad():
        enc_out = model.encode(src, src_mask)

    # Start with SOS
    ys = torch.tensor([[sos_idx]], dtype=torch.long, device=device)

    for _ in range(max_len - 1):
        tgt_mask = (ys != tokenizer_tgt.token_to_id("[PAD]")).unsqueeze(0).int() & causal_mask(ys.size(1)).to(device)
        dec_out = model.decode(ys, enc_out, src_mask, tgt_mask)
        logits = model.project(dec_out)  # (B=1, seq_len, vocab_size)
        next_token = torch.argmax(logits[:, -1], dim=-1).unsqueeze(0)  # last timestep
        ys = torch.cat([ys, next_token], dim=1)
        if next_token.item() == eos_idx:
            break
    return ys.squeeze(0)

@torch.no_grad()
def run_validation(model, val_dataloader, tokenizer_tgt, loss_fn, device):
    model.eval()
    total_loss = 0
    total_correct = 0
    total_tokens = 0

    # Compute average validation loss and accuracy
    for batch in val_dataloader:
        encoder_input = batch["encoder_input"].to(device)
        decoder_input = batch["decoder_input"].to(device)
        encoder_mask = batch["encoder_mask"].to(device)
        decoder_mask = batch["decoder_mask"].to(device)
        label = batch["label"].to(device)

        enc_out = model.encode(encoder_input, encoder_mask)
        dec_out = model.decode(decoder_input, enc_out, encoder_mask, decoder_mask)
        logits = model.project(dec_out)

        loss = loss_fn(logits.view(-1, logits.size(-1)), label.view(-1))
        total_loss += loss.item()

        # Calculate token-level accuracy (ignoring PAD tokens)
        preds = logits.argmax(dim=-1)
        mask = label != tokenizer_tgt.token_to_id("[PAD]")
        correct = (preds == label) & mask
        total_correct += correct.sum().item()
        total_tokens += mask.sum().item()

    avg_loss = total_loss / len(val_dataloader) if len(val_dataloader) > 0 else 0.0
    accuracy = total_correct / total_tokens if total_tokens > 0 else 0.0

    # Print 2 examples from the val set and show attention from the first encoder layer
    print("[INFO] Validation sample predictions:")
    max_examples = 2
    val_iter = iter(val_dataloader)
    for i in range(max_examples):
        try:
            batch = next(val_iter)
        except StopIteration:
            break

        src = batch["encoder_input"].to(device)
        src_mask = batch["encoder_mask"].to(device)
        lbl_text = batch["tgt_text"][0]  # ground truth

        # Greedy decode
        pred_ids = greedy_decode(model, src, src_mask, tokenizer_tgt, max_len=50, device=device)
        pred_text = tokenizer_tgt.decode(pred_ids.tolist())

        print(f"  Example {i+1}")
        print(f"    Source:   {batch['src_text'][0]}")
        print(f"    Expected: {lbl_text}")
        print(f"    Predicted:{pred_text}")
        print("    ------")
        
        # For the first example, plot attention from the first encoder layer's self-attention
        # (Assumes the first encoder layer's MultiHeadAttention stores its attention scores)
        enc_input = src
        _ = model.encode(enc_input, src_mask)
        # Access the attention scores from the first encoder layer
        attn_scores = model.encoder.layers[0].self_attention_block.attention_scores
        # Get tokens for the encoder (for plotting, we use the src tokens)
        tokens = [str(t) for t in enc_input[0].tolist()]
        plot_attention(attn_scores[0], tokens, title="Encoder Self-Attention (Layer 1)")

    return avg_loss, accuracy

# ---------------------------
# Training code
# ---------------------------
def train_model(config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"[INFO] Training on device: {device}")

    # Make weights folder
    folder = f"{config['datasource'].replace('/', '_')}_{config['model_folder']}"
    Path(folder).mkdir(exist_ok=True, parents=True)

    # Prepare data
    train_dataloader, val_dataloader, tokenizer_src, tokenizer_tgt = get_ds(config)
    src_vocab_size = tokenizer_src.get_vocab_size()
    tgt_vocab_size = tokenizer_tgt.get_vocab_size()

    # Build model
    model = build_transformer(
        src_vocab_size,
        tgt_vocab_size,
        config["seq_len"],
        config["seq_len"],
        d_model=config["d_model"]
    ).to(device)

    optimizer = optim.Adam(model.parameters(), lr=config["lr"], eps=1e-9)

    # Use target tokenizer's PAD as ignore_index
    loss_fn = nn.CrossEntropyLoss(
        ignore_index=tokenizer_tgt.token_to_id("[PAD]"), 
        label_smoothing=0.1
    ).to(device)

    # Optionally resume
    initial_epoch = 0
    global_step = 0
    model_filename = None
    if config["preload"] == "latest":
        model_filename = latest_weights_file_path(config)
    elif config["preload"]:
        model_filename = get_weights_file_path(config, config["preload"])

    if model_filename and Path(model_filename).exists():
        print(f"[INFO] Resuming from: {model_filename}")
        state = torch.load(model_filename, map_location=device)
        model.load_state_dict(state["model_state_dict"])
        optimizer.load_state_dict(state["optimizer_state_dict"])
        initial_epoch = state["epoch"] + 1
        global_step = state["global_step"]
    else:
        print("[INFO] Training from scratch.")

    # Lists for logging loss and accuracy
    train_losses_per_epoch = []
    val_losses_per_epoch = []
    train_acc_per_epoch = []
    val_acc_per_epoch = []
    times_per_epoch = []

    # Training loop
    for epoch in range(initial_epoch, config["num_epochs"]):
        model.train()
        torch.cuda.empty_cache()
        batch_bar = tqdm(train_dataloader, desc=f"Epoch {epoch}")

        start_time = time.time()
        total_train_loss = 0.0
        total_correct = 0
        total_tokens = 0
        batch_count = 0

        for batch in batch_bar:
            encoder_input = batch["encoder_input"].to(device, non_blocking=True)
            decoder_input = batch["decoder_input"].to(device, non_blocking=True)
            encoder_mask = batch["encoder_mask"].to(device, non_blocking=True)
            decoder_mask = batch["decoder_mask"].to(device, non_blocking=True)
            label = batch["label"].to(device, non_blocking=True)

            optimizer.zero_grad(set_to_none=True)

            enc_out = model.encode(encoder_input, encoder_mask)
            dec_out = model.decode(decoder_input, enc_out, encoder_mask, decoder_mask)
            logits = model.project(dec_out)

            loss = loss_fn(logits.view(-1, tgt_vocab_size), label.view(-1))
            loss.backward()
            optimizer.step()

            total_train_loss += loss.item()
            batch_count += 1

            # Calculate token-level accuracy for this batch (ignoring PAD tokens)
            preds = logits.argmax(dim=-1)
            mask = label != tokenizer_tgt.token_to_id("[PAD]")
            correct = (preds == label) & mask
            total_correct += correct.sum().item()
            total_tokens += mask.sum().item()

            global_step += 1
            batch_bar.set_postfix({"loss": f"{loss.item():.3f}"})

        # End epoch
        epoch_time = time.time() - start_time
        times_per_epoch.append(epoch_time)

        avg_train_loss = total_train_loss / (batch_count if batch_count else 1)
        train_losses_per_epoch.append(avg_train_loss)
        train_accuracy = total_correct / total_tokens if total_tokens > 0 else 0.0
        train_acc_per_epoch.append(train_accuracy)

        # Run validation
        val_loss, val_accuracy = run_validation(model, val_dataloader, tokenizer_tgt, loss_fn, device)
        val_losses_per_epoch.append(val_loss)
        val_acc_per_epoch.append(val_accuracy)

        print(f"[INFO] Epoch {epoch} - Train Loss: {avg_train_loss:.4f}, Train Acc: {train_accuracy:.4f}, "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy:.4f}, Time: {epoch_time:.2f}s")

        # Save checkpoint
        ckpt_path = get_weights_file_path(config, f"{epoch:02d}")
        torch.save({
            "epoch": epoch,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
            "global_step": global_step
        }, ckpt_path)
        print(f"[INFO] Epoch {epoch} finished, saved checkpoint to {ckpt_path}")

    # ---- Plot Training & Validation Loss ----
    plt.figure(figsize=(8, 5))
    plt.plot(range(initial_epoch, initial_epoch + config["num_epochs"]), train_losses_per_epoch, label="Train Loss", marker="o")
    plt.plot(range(initial_epoch, initial_epoch + config["num_epochs"]), val_losses_per_epoch, label="Val Loss", marker="s")
    plt.title("Training & Validation Loss over Epochs")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

    # ---- Plot Training & Validation Accuracy ----
    plt.figure(figsize=(8, 5))
    plt.plot(range(initial_epoch, initial_epoch + config["num_epochs"]), train_acc_per_epoch, label="Train Accuracy", marker="o")
    plt.plot(range(initial_epoch, initial_epoch + config["num_epochs"]), val_acc_per_epoch, label="Val Accuracy", marker="s")
    plt.title("Training & Validation Accuracy over Epochs")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    # ---- Print Times per Epoch ----
    for i, ttime in enumerate(times_per_epoch):
        print(f"Epoch {i + initial_epoch} took {ttime:.2f} seconds.")

# ---------------------------
# Run the training
# ---------------------------
if __name__ == "__main__":
    config = get_config()
    train_model(config)
```
